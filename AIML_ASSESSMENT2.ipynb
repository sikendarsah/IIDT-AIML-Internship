#**Missing Value Handling**

```python
import numpy as np
import pandas as pd
```

```python
# Sample DataFrame with missing values
data = {
    'A': [1, np.nan, 3, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': [1, 2, np.nan, 4, 5]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

```python
# Method 1: Removing missing values
# Remove rows with any missing values
cleaned_df1 = df.dropna()

print("\nCleaned DataFrame after method 1 (removing missing values):")
print(cleaned_df1)
```

```python
# Method 2: Imputation
# Fill missing values with a specific constant
cleaned_df2 = df.fillna(0)

print("\nCleaned DataFrame after method 2 (imputation with constant):")
print(cleaned_df2)
```

```python
# Method 3: Imputation using mean, median, or mode
# Fill missing values with mean of respective columns
cleaned_df3 = df.fillna(df.mean())


print("\nCleaned DataFrame after method 3 (imputation with mean):")
print(cleaned_df3)
```

```python

# Method 4: Forward fill or backward fill
# Fill missing values with the nearest known value in the respective column
cleaned_df4 = df.ffill()  # forward fill
# cleaned_df4 = df.bfill()  # backward fill


print("\nCleaned DataFrame after method 4 (forward fill or backward fill):")
print(cleaned_df4)
```

```python
# Method 5: Interpolation
# Fill missing values by interpolating between neighboring values
cleaned_df5 = df.interpolate()


print("\nCleaned DataFrame after method 5 (interpolation):")
print(cleaned_df5)
```

#**Outlier Handling**

```python
import numpy as np
from scipy import stats
```

```python
# Generate some sample data
np.random.seed(0)
data = np.random.normal(0, 1, 1000)
```

```python
# Define a threshold for identifying outliers (e.g., Z-score > 3)
threshold = 3
```

```python
# Calculate Z-scores for each data point
z_scores = np.abs(stats.zscore(data))
```

```python
# Identify outliers
outliers = np.where(z_scores > threshold)[0]
```

```python
# Optionally remove outliers from the dataset
cleaned_data = data[~np.isin(np.arange(len(data)), outliers)]
```

```python

print("Identified outliers indices:", outliers)
print("Cleaned data shape:", cleaned_data.shape)
```

#**Irrelevant Data Handling**

```python
import pandas as pd
```

```python
# Sample DataFrame with irrelevant columns
data = {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami']
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

```python
# Method 1: Dropping irrelevant columns
# Define a list of irrelevant columns to drop
irrelevant_columns = ['ID', 'Name']
cleaned_df1 = df.drop(columns=irrelevant_columns)

print("\nCleaned DataFrame after method 1 (dropping irrelevant columns):")
print(cleaned_df1)
```

```python
# Method 2: Manual selection of relevant columns
# Specify only the relevant columns to keep
relevant_columns = ['Age', 'Salary', 'City']
cleaned_df2 = df[relevant_columns]

print("\nCleaned DataFrame after method 2 (manual selection of relevant columns):")
print(cleaned_df2)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load sample dataset (Iris dataset for demonstration)
iris = sns.load_dataset('iris')
```

```python
# Display basic information about the dataset
print("Basic information about the dataset:")
print(iris.info())
```

```python
# Summary statistics
print("\nSummary statistics of numerical features:")
print(iris.describe())
```

```python
# Visualization: Pairplot for pairwise relationships between features
sns.pairplot(iris, hue='species', markers=['o', 's', 'D'])
plt.title("Pairplot of Iris dataset")
plt.show()
```

```python
# Visualization: Boxplot for each feature by species
plt.figure(figsize=(10, 6))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=feature, data=iris)
    plt.title(f"{feature} by Species")
plt.tight_layout()
plt.show()
```

#**Feature Scaling**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.datasets import load_iris
```

```python
# Load sample dataset (Iris dataset for demonstration)
iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
```

```python
# Display original dataset
print("Original dataset:")
print(iris.head())
```

```python
# Method 1: Standardization (Z-score normalization)
scaler_standard = StandardScaler()
iris_standardized = scaler_standard.fit_transform(iris)
# Display standardized dataset
print("\nStandardized dataset (Z-score normalization):")
print(pd.DataFrame(iris_standardized, columns=iris.columns).head())
```

```python
# Method 2: Min-Max Scaling
scaler_minmax = MinMaxScaler()
iris_minmax = scaler_minmax.fit_transform(iris)

# Display Min-Max scaled dataset
print("\nMin-Max scaled dataset:")
print(pd.DataFrame(iris_minmax, columns=iris.columns).head())
```

```python

# Method 3: Robust Scaling
scaler_robust = RobustScaler()
iris_robust = scaler_robust.fit_transform(iris)

# Display Robust scaled dataset
print("\nRobust scaled dataset:")
print(pd.DataFrame(iris_robust, columns=iris.columns).head())
```

#**Categorical Encoding Operations with Titanic Dataset**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.datasets import load_iris
```

```python
# Load sample dataset (Iris dataset for demonstration)
iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['species'] = iris_data.target

# Display original dataset
print("Original dataset:")
print(iris.head())
```

```python
# Method 1: Label Encoding
label_encoder = LabelEncoder()
iris_label_encoded = iris.copy()
iris_label_encoded['species_encoded'] = label_encoder.fit_transform(iris['species'])

# Display dataset with label encoded categorical variable
print("\nDataset with Label Encoded categorical variable:")
print(iris_label_encoded.head())
```

```python

# Method 2: One-Hot Encoding
onehot_encoder = OneHotEncoder(sparse=False)
species_onehot = onehot_encoder.fit_transform(iris[['species']])

# Create DataFrame from one-hot encoded array
iris_onehot = pd.concat([iris, pd.DataFrame(species_onehot, columns=['species_{}'.format(i) for i in range(species_onehot.shape[1])])], axis=1)

# Drop original 'species' column
iris_onehot.drop('species', axis=1, inplace=True)

# Display dataset with one-hot encoded categorical variable
print("\nDataset with One-Hot Encoded categorical variable:")
print(iris_onehot.head())
```
